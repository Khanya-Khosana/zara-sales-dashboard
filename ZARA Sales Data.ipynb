```python
mode_va = sales['name'].mode()[0]
mode_va
```

```python
#converting boolean column 
sales['Promotion'] = sales['Promotion'].astype('category')
sales['Seasonal'] = sales['Seasonal'].astype('category')

#converting categorical column
sales['Product Category'] = sales['Product Category'].astype('category')
sales['section'] = sales['section'].astype('category')
sales['brand'] = sales['brand'].astype('category')
sales['currency'] = sales['currency'].astype('category')
# convert to proper date
sales['scraped_at'] = pd.to_datetime(sales['scraped_at'])
```

```python
 # Strip whitespace and fix casing
sales['Product Category'] = sales['Product Category'].str.strip().str.title()
sales['section'] = sales['section'].str.strip().str.lower()
sales['brand'] = sales['brand'].str.strip().str.title()
```

```python
# Revenue column
sales['Revenue'] = sales['price'] * sales['Sales Volume']

# Add month for time-based grouping
sales['Scraped Month'] = sales['scraped_at'].dt.to_period('M')
```

```python
sales.drop(columns=['url', 'sku', 'terms'], inplace=True)
```
